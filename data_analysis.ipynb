```python
import pandas as pd
import numpy as np
import seaborn as sns

def get_crosstab(df, feature):
    crosstab = pd.crosstab(df[feature], df['is_converted'], normalize='index')*100
    crosstab = crosstab.reset_index()
    return crosstab

def na_percentage(df, feature):
    percentage = (df[feature].isna().sum() / row_num) * 100
    return percentage

df_train = pd.read_csv('./train.csv')
row_num = len(df_train)
```

```python
print("결측치 존재 요소와 요소 별 결측치 비율")
print(f"customer_country : 약 {na_percentage(df_train, 'customer_country') : .2f} %", )
print(f"com_reg_ver_win_rate : 약 {na_percentage(df_train, 'com_reg_ver_win_rate') : .2f} %", )
print(f"customer_type : 약 {na_percentage(df_train, 'customer_type') : .2f} %", )
print(f"historical_existing_cnt : 약 {na_percentage(df_train, 'historical_existing_cnt') : .2f} %", )
print(f"id_strategic_ver : 약 {na_percentage(df_train, 'id_strategic_ver') : .2f} %", )
print(f"it_strategic_ver : 약 {na_percentage(df_train, 'it_strategic_ver') : .2f} %", )
print(f"idit_strategic_ver : 약 {na_percentage(df_train, 'idit_strategic_ver') : .2f} %", )
print(f"customer_job : 약 {na_percentage(df_train, 'customer_job') : .2f} %", )
print(f"inquiry_type : 약 {na_percentage(df_train, 'inquiry_type') : .2f} %", )
print(f"product_category : 약 {na_percentage(df_train, 'product_category') : .2f} %", )
print(f"product_subcategory : 약 {na_percentage(df_train, 'product_subcategory') : .2f} %", )
print(f"product_modelname : 약 {na_percentage(df_train, 'product_modelname') : .2f} %", )
print(f"customer_country : 약 {na_percentage(df_train, 'customer_country') : .2f} %", )
print(f"expected_timeline : 약 {na_percentage(df_train, 'expected_timeline') : .2f} %", )
print(f"ver_win_rate_x : 약 {na_percentage(df_train, 'ver_win_rate_x') : .2f} %", )
print(f"ver_win_ratio_per_bu : 약 {na_percentage(df_train, 'ver_win_ratio_per_bu') : .2f} %", )
print(f"business_area : 약 {na_percentage(df_train, 'business_area') : .2f} %", )
print(f"business_subarea : 약 {na_percentage(df_train, 'business_subarea') : .2f} %", )
```

```python
over_true_200 = true_converted_lead[true_converted_lead.values > 200]
true_over_per = len(over_true_200) / true_num
true_under_per = 1 - true_over_per
print(true_over_per) # is_converted 값이 True 인 것 중 lead_desc_length 가 200 보다 클 확률
print(true_under_per) # is_converted 값이 True 인 것 중 lead_desc_length 가 200 보다 작거나 같을 확률
```

```python
over_false_200 = false_converted_lead[false_converted_lead.values > 200]
false_over_per = len(over_false_200) / false_num
false_under_per = 1 - false_over_per
print(false_over_per) # is_converted 값이 False 인 것 중 lead_desc_length 가 200 보다 클 확률
print(false_under_per) # is_converted 값이 False 인 것 중 lead_desc_length 가 200 보다 작거나 같을 확률
```
